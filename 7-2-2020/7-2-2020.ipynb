## STEPS

```python
##    ----->Data calling
####  ----->x,y
####  ----->from sk learn model calling
###    ----->train,test
##    ----->Data fitting to model
###  ----->predict[testing]
###    ----->score[train,testing]
###    ----->plotting
```

```python
boston.data
```

```python
boston.target
```

```python
boston.feature_names
```

```python
## Creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["Land Rate"]=boston.target
```

```python
boston_df.head()
```

```python
x=boston_df.drop("Land Rate",axis=1)
xy=
```

```python
x.values
```

```python
y=boston_df["Land Rate"]
y.values
```

```python
#Ridge Regression=sum of error+sum of squares of magnitude of coefficient(alpha)
##coefficient(alpha)
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
rr
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
rr.fit(x_train,y_train)
print(x_train)
```

```python
y_predict=rr.predict(x_train)
```

```python
y_predict
```

```python
#Score 
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
lintrain_score=lin.score(x_train,y_train)
print("LIn Train Score",lintrain_score)
```

```python
lintrain_score=lin.score(x_test,y_test)
print("LIn Test Score",lintrain_score)
```

```python
lin.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lin.coef_,marker=".",label="Linear Coeffi")
plt.xlabel("coefficient index")
plt.ylabel("cpefficient magnitude")
plt.title("Ridge Vs Linear")
plt.legend()
plt.show()
```
