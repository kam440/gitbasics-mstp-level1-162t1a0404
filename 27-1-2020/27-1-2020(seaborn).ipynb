```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('tips')
t.head()
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('tips')
t.tail()
```

```python
sns.relplot(x="total_bill",y="day",data=t,hue="tip")
```

```python
sns.relplot(x="total_bill",y="size",data=t,hue="tip",kind="line")
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('flights')
t.head()
```

```python
sns.set(style="darkgrid")
f=sns.load_dataset('flights')
f.tail()
```

```python
sns.relplot(x="year",y="month",hue="passengers",data=f)
```

```python
sns.relplot(x="year",y="passengers",hue="month",col="month",data=f)
```

```python
sns.set(style="whitegrid")
i=sns.load_dataset("iris")
i.head()
```

```python
i["species"].value_counts()
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i)
```

```python
sns.relplot(x="sepal_width",y="petal_width",hue="species",
            col="species",data=i)
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('tips')
t.tail()
```

```python
sns.relplot("total_bill","tip",data=t,style="smoker",size="size",col="time",legend="full")
```

```python
## linear regression model plot
sns.lmplot("total_bill","tip",data=t)
```

```python
sns.lmplot("total_bill","tip",data=t,hue="smoker",col="time")
```

```python
# categorial data plot
## catplot
   ##swarm plot
    ## scatter plot
    ## bar plot
```

```python
sns.catplot(x="species",y="petal_length",data=i)
```

```python
sns.catplot(x="species",y="petal_length",data=i,hue="sepal_width")
```

```python
sns.catplot(x="species",y="petal_length",data=i,hue="sepal_width",kind="swarm")
```

```python
sns.catplot(x="species",y="petal_length",data=i,kind="violin")
```

```python
sns.barplot(x="species",y="petal_length",data=i,hue="sepal_width")
```

```python
sns.barplot(x="species",y="petal_length",data=i,hue="sepal_width")
```

```python
## box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",hue="petal_length",data=i,ax=axs[1])
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset('tips')
t.tail()
```

```python
sns.jointplot(x="total_bill",y="tip",data=t,kind="kde",hue="smoker")
```

```python
sns.jointplot(x="total_bill",y="tip",
              data=t,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",
              data=t,kind="reg",color="green",height=3.5,ratio=5)
```

```python
##pairplot
sns.pairplot(data=i,hue="species",palette="muted")
```

```python
sns.pairplot(data=i,hue="species",markers=["x","s","d"])
```

```python
sns.pairplot(i,vars=["sepal_width","sepal_length","petal_width","petal_length"])
```

```python
sns.pairplot(i,diag_kind="kde",markers="+")
```

```python
## Dist plot
x=np.random.rand(50)
sns.distplot(x)
```

```python
tt=sns.FacetGrid(i,col="species")
tt.map(plt.hist,"sepal_length")
```
