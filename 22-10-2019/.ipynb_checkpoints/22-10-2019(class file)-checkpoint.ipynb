## By using range in line plot

```python
import matplotlib.pyplot as p_l
x=list(range(1,10,2))  #[expresson loop condition]
y=list(range(20,30,2))
p_l.plot(x,y,'ro')
p_l.show()
```

```python
for i in range(1,10,2):
    print(i)
for j in range(20,30,2):
    print(j)
```

```python
# List comprehension
l=[i for i in range(1,10,2)]
l
```

```python
# Lineplot by using list comprehension
x=[i for i in range(1,11)]
print(l)
l=[i**0.5 for i in range(1,11)]
print(l)
p.plot(x,l,'ys')
p.grid()
```

## Labels names and title for a graph
### lib.xlabel('x-axis')
### lib.ylabel('y-axis')
### lib.title('sqrt of 1-10')

```python
import matplotlib.pyplot as h
x=[2,4]
y=[2,0]
h.plot(x,y)
h.xlabel('x-Data')
h.ylabel('y-Axis')
h.title('line plot')
h.grid()
h.show()
```

```python
# draw a plot for time and distance by using list comprehension
#  distance=2*time
import matplotlib.pyplot as p1
x=[i for i in range(1,10,2)]
print(x)
y=[i*2 for i in range(1,10,2)]
p1.xlabel('x-Time')
p1.ylabel('y-Distance')
p1.title('distance vs time')
p1.plot(x,y)
p1.show()
```

# Representing a sine wave
## A=sine(wt)
## A=Amplitude
## w=frequency
## t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title('Sine wave A=sin(wt)')
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2:]
Alaska_income=df.iloc[1,2:]
pl.plot(years,Alaska_income)
pl.title('Income of Alaska 2005-2013')
pl.xlabel('Years')
pl.savefig('Income of Alaska 2005-2013.jpg')
pl.ylabel('Income of Alaska')
```

# Plot for all states income from 2005-2013


```python
import matplotlib.pyplot as pl
import pandas as pd
years=df.columns[-5::]
income=df.iloc[0:,2:]
pl.plot(years,income)
pl.title('All States Income')
pl.xlabel('Years')
pl.ylabel('States')
pl.savefig('All States Income.jpg')
pl.show()
```
